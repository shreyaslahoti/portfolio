```r
#library(readxl)
library(ggplot2)
library(scales)
#library(ggrepel)

IPEDS_data <- read_excel("IPEDS_data.xlsx", sheet = 1) 
colnames(IPEDS_data) <- make.names(names(IPEDS_data))

bubble <- ggplot(data= IPEDS_data, aes(x=Percent.of.freshmen.receiving.any.financial.aid/100,y= Graduation.rate...Bachelor.degree.within.5.years..total,fill=Profile))+
geom_point(aes(size=Sector.of.institution),alpha=0.5,shape=21)+labs(x = " Percent of freshmen receiving any financial aid ", y = "Graduation Rate (Bachelor's degree within 5 Years, total) ", title = "Graduation rate - Bachelors degree within 5 years, total and Percent of Freshmen receiving any financial aid (size on sector of institution, color by Profile")+
scale_x_continuous(label=percent)

print(bubble)
```

```r
Densityplot <- ggplot(data=IPEDS_data,aes(x=Admissions.yield...total ))
Densityplot+ geom_density(aes(group= Profile, fill=Profile),adjust=2, alpha=0.5)+
labs(x = " Admissions Yield ",title = "Admissions yield by Profile")+
  scale_x_continuous()
```
